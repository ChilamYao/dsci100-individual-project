---

This project uses two datasets from a Minecraft research server.

The `players.csv` file has information about 196 players with 7 columns:
* `Age`: How old each player is (in years). Two players didn't provide their age.
* `played_hours`: How many hours each player spent on the server in total.
* `experience`: How much gaming experience players say they have.
* `gender`: The player's gender.
* `subscribe`: Whether the player signed up for the game newsletter (this is what we want to predict).
* `hashedEmail`: A coded ID number for each player that keeps their email private. We use this to connect the two datasets.
* `name`: The player's username (we won't use this for predictions).

The `sessions.csv` file has 1,535 rows with information about each time someone played:
* `hashedEmail`: Shows which player had each gaming session.
* `start_time` and `end_time`: When each gaming session started and ended. Two sessions are missing end times.
* `original_start_time` and `original_end_time`: The same times but shown as numbers. Two sessions are missing the ending number.

Each row in `players` shows one person, while each row in `sessions` shows one gaming session. Later, we'll combine the session data for each player to create new information, like counting how many times each player logged in.

Looking at the numbers, players are on average 21.14 years old (we have ages for 194 players), and they've played an average of 5.85 hours total.

<span style="font-size:30px;">(2) Questions</span>

Broad question: What player characteristics and behaviours are most predictive of subscribing to the game-related newsletter, and how do these features differ between various player types?
Specific question: Can we use player information like age, gender, gaming experience, and how long they've played to predict whether they'll subscribe to the newsletter?
This question matters because subscribing to the newsletter shows that players are interested and engaged with the project. If we can figure out which types of players are more likely to subscribe, the research team can focus their recruiting efforts on similar players and better understand their most active community members.

<span style="font-size:30px;">(3) Exploratory Data Analysis and Visualization</span>

```r
# Histogram
ggplot(players, aes(x = played_hours, fill = subscribe)) +
  geom_histogram(binwidth = 5, colour = "black") +
  labs(
    title = "Hours Played by Subscription",
    x = "Hours Played",
    y = "Number of Players",
    fill = "Subscribed"
  )

# Scatter
ggplot(players, aes(x = Age, y = played_hours, colour = subscribe)) +
  geom_point(alpha = 0.5, size = 2) +
  labs(
    title = "Age vs Hours Played",
    x = "Age",
    y = "Hours Played",
    colour = "Subscribed"
  )

# Bar by gender
players |>
  group_by(gender) |>
  summarise(subscribe_rate = mean(subscribe)) |>
  ggplot(aes(x = gender, y = subscribe_rate, fill = gender)) +
  geom_col() +
  labs(
    title = "Subscribe Rate by Gender",
    x = "Gender",
    y = "Subscribe Rate"
  ) +
  theme(axis.text.x = element_text(angle = 60, hjust = 1))
```
